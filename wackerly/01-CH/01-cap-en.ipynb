### **Exercise 1.1**
#### **Statement:**

For each of the following situations, identify the population of interest, the inferential objective, and how you might go about collecting a sample.

1. The National Highway Safety Council wants to estimate the proportion of automobile tires with unsafe tread among all tires manufactured by a specific company during the current production year.

2. A political scientist wants to determine whether a majority of adult residents of a state favor a unicameral legislature.

3. A medical scientist wants to estimate the average length of time until the recurrence of a certain disease.

4. An electrical engineer wants to determine whether the average length of life of transistors of a certain type is greater than 500 hours.

5. A university researcher wants to estimate the proportion of U.S. citizens from "Generation X" who are interested in starting their own business.

6. For more than a century, normal body temperature for humans has been accepted to be 98.6 Fahrenheit. Is it really? Researchers want to estimate the average temperature of healthy adults in the United States.

7. A city engineer wants to estimate the average weekly water consumption for single-family dwelling units in the city.

(Wackerly, Mendenhall & Scheaffer, 2008, p. 8)

#### **Solution**
**Theory Section:** \
We identify the 3 parts necessary for our solution, *population of interest*, *the inferential objective*, *idea on collecting sample from population*. 


1. The population of interest is all the tires fabricated by the specific company for this specific production year. The inferential objective is to predict the total number of tires with unsafe tread manufactured during the current production year. A sample could be collected by taking random tires from the assembly line and testing the safety of the tread.

2. The population of interest is adult residents of a state. The inferential objective is to determine wether a majority of the population is in favor of an unicameral legislature. To collect a sample there should be brigades of people conducting door to door pools, or cellphone calls by area registration or ads asking for an answer to the poll for social media users that live in that state.

3. The population of interest would be the patients that approach certain clinics with symptoms of that disease. The inferential objective is the rate at which a person appears in the system with that disease and further down the analysis the rate at which people get sick with this disease. A sample could be collected by nurses and medical techs filling statistical report sheets designed to detect symptoms of patients from hospitals and medical centers of the country selected as representative for the experiment.

4. The population of interest is the set of all transistors of a certain type. The inferential objective is the average length of life of the transistors of a certain type. To take a sample one would have to create a standard for transistors of certain type so that all manufacturers that build them put them under this test.

5. The population of interest is the U.S. citizens from "Generation X". The inferential objective is to find the proportion of "gen X" U.S. citizens interested in starting their own business. Work with banks to obtain the contact information of "Gen X" citizens and interview them.

6. The population of interest is the set of all healthy adults in the U.S. The inferential objective is to estimate the average temperature of the population using samples of healthy adults. Samples could be taken by using advertisements in health apps to determine which adults are healthy and then have them fill a form for data.

7. The population of interest are all families that consume water inside the city system. The inferential objective is the weekly rate at which water is consumed by families in the city. A sample could be taken by accessing public records and studying consumption for each service charge.

  This is obtained by looking at the histogram and obtaining the values of intersection.

3. The categories that have more than 0 but fewer than 5 quarts are $C = \{1,2,3,4 \} $. Interpreting the histogram we have:
\begin{align*}
\mathrm{Proportion}(0<\mathrm{Quarts}<5) = 0.2 + 0.36 + 0.2 + 0.12 = 0.88
\end{align*}

### **Exercise 1.9**
#### **Statement:**

Resting breathing rates for college-age students are approximately normally distributed with mean 12 and standard deviation 2.3 breaths per minute. What fraction of all college-age students have breathing rates in the following intervals?

1. 9.7 to 14.3 breaths per minute.

2. 7.4 to 16.6 breaths per minute.

3. 9.7 to 16.6 breaths per minute.

4. Less than 5.1 or more than 18.9 breaths per minute.

(Wackerly, Mendenhall & Scheaffer, 2008, p. X)

#### **Solution**
**Theory Section:**

1. We find that $9.7 + 2.3 = 12$ and $14.3 - 2.3 = 12$, this creates an interval with endpoints $ \mu \pm \sigma$ which contains approximately 68% of the measurements.

2. We find that $7.4 + 2(2.3) = 12$ and $16.6 - 2(2.3) = 12$, this creates an interval with endpoints $ \mu \pm 2\sigma $ which contains approximately 95% of the measurements.

3. Having $9.7 + 2.3 = 12$ and $16.6 - 2(2.3) = 12$, this creates the interval with endpoints $\left[\mu-\sigma, \mu + 2\sigma \right]$, this interval contains is the union of intervals $\left[\mu-\sigma,\mu\right]$ and $\left[\mu, \mu + 2\sigma\right]$. The first fractionate interval ($\left[\mu-\sigma,\mu\right]$) contains half of the interval with endpoints $\mu \pm \sigma$. The second interval ($\left[\mu, \mu + 2\sigma\right]$) contains half of the interval with endpoints $\mu \pm 2\sigma$. The addition of the proportion follows:

\begin{align*}
\mathrm{Proportion} ([ \mu -\sigma, \mu + 2 \sigma ] ) = \frac{\mathrm{Proportion} ( \left[\mu - \sigma,\mu + \sigma \right])}{2} + \frac{ \mathrm{Proportion} ( \left[\mu - 2 \sigma,\mu + 2 \sigma \right] ) }{2} = \frac{68\%}{2} + \frac{95\%}{2} = 34\% + 47.5 \% = 81.5\%
\end{align*}


&emsp;&emsp;&ensp;This is an approximation, the exact value might be different. The region under the  normal distribution and its proportion of the total measurements is shown in the code section.


4. For this interval, we have the union of intervals $\left(- \infty, \mu - 3\sigma \right] \bigcup \left[\mu + 3 \sigma, \infty \right)$, which is outside the range for most measurements and very small. An exact value is calculated in the code below.

**Code:**

```python
# Definition
mu = 12         
sigma = 2.3      
dist = norm(loc=mu, scale=sigma)

# Values of the probability function
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y = dist.pdf(x)    

# ==========================> Answer to 3. <===============================

# Figure and ax definition
fig6, ax6 = plt.subplots(figsize=(8, 3))
ax6.plot(x, y, color="black", lw=2)

# Vertical lines
for k in [-3, -2, -1, 0, 1, 2, 3]:
    ax6.axvline(mu + k*sigma, color="black", lw=1)

# Fill area under graph
mask = (x >= mu - sigma) & (x <= mu + 2*sigma)
ax6.fill_between(x[mask], y[mask], alpha=0.3)

# Visuals
ax6.set_ylim(0, y.max()*1.1)
ax6.set_xticks([mu + 2.3*sigma for sigma in np.linspace(-3, 3,num = 7)])
ax6.set_xticklabels([r"$\mu - 3\sigma$", r"$\mu - 2\sigma$", r"$\mu - \sigma$", r"$\mu$", r"$\mu + \sigma$", r"$\mu + 2\sigma$", r"$\mu + 3\sigma$"])
ax6.set_yticks([])

print("For the interval in 3, the proportion is:")
plt.show()

prop = norm.cdf(mu + 2*sigma, mu, sigma) - norm.cdf(mu - sigma, mu, sigma)
print(f"The proportion of the measurements in 3 is {prop*100:.2f}%")
print()

# ==========================> Answer to 4. <===============================

prop2 = 2*norm.cdf(mu-3*sigma, mu, sigma)
print(f"The proportion of the measurements in 4 is {prop2*100:.2f}%")
```

### **Exercise 1.11**
#### **Statement:**

The following results on summations will help us in calculating the sample variance $s^{2}$. For any constant $c$.

1. $\sum_{i=1}^{n}c=nc$.

2. $\sum_{i=1}^{n}cy_{i} = c \sum_{i=1}^{n}y_{i}$.

3. $\sum_{i=1}^{n}(x_{i}+y_{i})=\sum_{i=1}^{n}y_{i}^{2} + \sum_{i=1}^{n}y_{i}$.

Use (1), (2) and (3) to show that

\begin{align*}
s^{2} = \frac{1}{n-1} \sum_{i=1}^{n}(y_{i} - \bar{y})^{2} = \frac{1}{n-1}\left[ \sum_{i=1}^{n}y_{i}^{2}-\frac{1}{n} \left( \sum_{i=1}^{n} y_{i} \right)^{2} \right]
\end{align*}

(Wackerly, Mendenhall & Scheaffer, 2008, p. X)

#### **Solution**
**Theory Section:**

To show that the equation holds, start from the left side

\begin{align*}
s^{2} = \frac{1}{n-1} \sum_{i=1}^{n}(y_{i} - \bar{y})^{2} =& \frac{1}{n-1} \sum_{i=1}^{n} \left( y_{i}^{2} - 2y_{i}\bar{y}+\bar{y}^{2} \right) \\
=& \frac{1}{n-1} \left[ \sum_{i=1}^{n} y_{i}^{2} +  \sum_{i=1}^{n} \left( - 2y_{i}\bar{y}+\bar{y}^{2} \right) \right]
\end{align*}

Using 3, then

\begin{align*}
\frac{1}{n-1} \left[ \sum_{i=1}^{n} y_{i}^{2} +  \sum_{i=1}^{n} \left( - 2y_{i}\bar{y}+\bar{y}^{2} \right) \right] =& \frac{1}{n-1} \left[ \sum_{i=1}^{n} y_{i}^{2} +  \bar{y}\sum_{i=1}^{n} \left( - 2y_{i}+\bar{y} \right) \right]
\end{align*}

By using 2, then using 3 again

\begin{align*}
\frac{1}{n-1} \left[ \sum_{i=1}^{n} y_{i}^{2} +  \bar{y}\sum_{i=1}^{n} \left( - 2y_{i}+\bar{y} \right) \right] =& \frac{1}{n-1} \left[ \sum_{i=1}^{n} y_{i}^{2} +  \bar{y} \left( - 2\sum_{i=1}^{n}y_{i}+ \sum_{i=1}^{n} \bar{y} \right) \right] \\
=& \frac{1}{n-1} \left[ \sum_{i=1}^{n} y_{i}^{2} +  \bar{y} \left( - 2\sum_{i=1}^{n}y_{i}+ n \bar{y} \right) \right]
\end{align*}

Because $\sum_{i=1}^{n}\bar{y} = n\bar{y}$, furthermore

\begin{align*}
\frac{1}{n-1} \left[ \sum_{i=1}^{n} y_{i}^{2} +  \bar{y} \left( - 2\sum_{i=1}^{n}y_{i}+ n \bar{y} \right) \right] =& \frac{1}{n-1} \left[ \sum_{i=1}^{n} y_{i}^{2} +  \frac{1}{n} \sum_{i=1}^{n} y_{i} \left( - 2\sum_{i=1}^{n}y_{i}+ \sum_{i=1}^{n} y_{i} \right) \right] \\
=& \frac{1}{n-1} \left[ \sum_{i=1}^{n} y_{i}^{2} -  \frac{1}{n} \left( \sum_{i=1}^{n} y_{i} \right)^{2} \right]
\end{align*}

Which reaches the right side of the equation to prove.

### **Exercise 1.12**
#### **Statement:**

Use the result of Exercise 1.11 to calculate $s$ for the $n=6$ sample measurements 1, 4, 2, 1, 3, and 3.

(Wackerly, Mendenhall & Scheaffer, 2008, p. X)

#### **Solution**
**Theory Section:**

Using python to compare different implementations that can be done for the variance calculation, first the classic definition, then the result of exercise 1.11 and finally using the numpy `np.var()` function with up to 4 decimal digits.

**Code:**

```python
# First define variance using the classic implementation

def varianceClassic(vector):
    var = 0
    mean = np.mean(vector)
    for y in vector:
        var += (y - mean) **2
    return var/(np.size(vector)-1)


# Secondly define variance by the result of exercise 1.11

def variance1_11(vector):
    sum1 = np.sum(vector*vector)
    sum2 = np.sum(vector)
    return (sum1 - (pow(sum2, 2))/(np.size(vector)))/(np.size(vector)-1)

# Finally we compute the data and the variances

data = np.array([1, 4, 2, 1, 3, 3])

print(f"The classic variance for this data is {varianceClassic(data):.4f}.")
print()
print(f"The variance using the result of exercise 1.11 is {variance1_11(data):.4f}.")
print()
print(f"The variance using numpy is {np.var(data, ddof = 1):.4f}.") # Here we use the ddof parameter for the variance to divide by n - 1 and not by n which is default.
```

### **Exercise 1.17**
#### **Statement:**

The *range* of a set of measurements is the difference between the largest and the smallest values. The empirical rule suggests that the standard deviation of a set of measurements may be roughly approximated by one-fourth of the range (that is, $\mathrm{range}/4$). Calculate this approximation to $s$ for the data sets in Exercises 1.2, 1.3, and 1.4. Compare the result in each case to the actual, calculated value of $s$. 

(Wackerly, Mendenhall & Scheaffer, 2008, p. 12)

#### **Solution**
**Theory Section:**

The solution to this exercise comes by calculates the defined approximation for the standard deviation of a sample by dividing its range by 4 ($s \approx \frac{\mathrm{range}}{4}$) and using the function `stdev()` from the `statistics` built int package.

**Code:**

```python
datasets = {
                "Ex 1.2": [8.9, 12.4, 8.6, 11.3, 9.2, 8.8, 35.1, 6.2, 7.0,
                            7.1, 11.8, 10.7, 7.6, 9.1, 9.2, 8.2, 9.0, 8.7,
                            9.1, 10.9, 10.3, 9.6, 7.8, 11.5, 9.3, 7.9, 8.8,
                            8.8, 12.7, 8.4, 7.8, 5.7, 10.5, 10.5, 9.6, 8.9,
                            10.2, 10.3, 7.7, 10.6, 8.3, 8.8, 9.5, 8.8, 9.4],
                
                "Ex 1.3": [0.74, 6.47, 1.90, 2.69, .75,
                            0.32, 9.99, 1.77, 2.41, 1.96,
                            1.66, 0.70, 2.42, 0.54, 3.36,
                            3.59, 0.37, 1.09, 8.32, 4.06,
                            4.55, 0.76, 2.03, 5.70, 12.48],

                "Ex 1.4": [11.88, 6.27, 5.49, 4.81, 4.40, 3.78, 3.44, 3.11, 2.88, 2.68,
                            7.99, 6.07, 5.26, 4.79, 4.05, 3.69, 3.36, 3.03, 2.74, 2.63,
                            7.15, 5.98, 5.07, 4.55, 3.94, 3.62, 3.26, 2.99, 2.74, 2.62,
                            7.13, 5.91, 4.94, 4.43, 3.93, 3.48, 3.20, 2.89, 2.69, 2.61]
            }

# Calculo de los rangos

def range_and_std(values):
    """
    Returns:
          - R/4 . The approximation of the standard deviation, where R is the range of values.
          - s . The value of the standard deviation of values.
    """
    data_range = max(values) - min(values)
    return data_range/4, stdev(values)

# Respuesta del ejercicio

for label, data in datasets.items():
    approx, exact = range_and_std(data)
    print(f"{label} R/4 = {approx:.2f}   s = {exact:.2f}")
```

### **Exercise 1.22**
#### **Statement:**

Prove that the sum of the deviations of a set of measurements about their mean is equal to zero; that is,

\begin{align*}
\sum_{i=1}^{n}\left( y_{i} - \bar{y} \right) = 0.
\end{align*}

(Wackerly, Mendenhall & Scheaffer, 2008, p. 16)

#### **Solution**
**Theory Section:**

First by distributing the sum over both $y_{i}$ and $\bar{y}$

\begin{align*}
\sum_{i=1}^{n}\left( y_{i} - \bar{y} \right) = \sum_{i=1}^{n}y_{i} - \sum_{i=1}^{n} \bar{y}
\end{align*}

Then, knowing that $n\bar{y} = \sum_{i=1}^{n}y_{i}$,

\begin{align*}
n\bar{y} - \sum_{i=1}^{n} \bar{y} = n\bar{y} - n\bar{y} = 0
\end{align*}

(Wackerly, Mendenhall & Scheaffer, 2008, p. 17, 18)

#### **Solution**
**Theory Section:**

First the statement that there are $48\%$ of measurements inside the interval $\left[ 2.5, 3.6 \right]$ is incorrect, given that this is a normal distribution $\bar{y} + 2s < 2.5$. From the theoretical approximation, the percentage of measurements found in the interval $\left[ \bar{y} + 2s, \infty \right)$ is $2.5\%$, so the stated percentage of $48\%$ of measurements for the interval $\left[ 2.5, 3.6 \right]$, which is a subset of $\left[ \bar{y} + 2s, \infty \right)$ contradicts both theory and reason. Using python in the code section we find the value found in a generated random normal sample of 2000 values and the probability for a population distribution with the mentioned parameters.

To find the probability that a measurement falls in the interval $\left[ 1.2, 2.4 \right]$ we can use 3 different methods.

- Using the theoretical approximation, the interval $\left[ 1.2, 2.4 \right]$ corresponds to the interval $\left[ \bar{y} + s, \bar{y} + 2s \right]$ which corresponds to a probability of $\frac{95 - 86}{2}\% = 13.5 \%$.

- Using the `cdf()` function from the `norm` module of `scipy.stats` for the interval $\left[ 1.2, 2.4 \right]$ we find that the probability is approximately $15\%$.

- Using the results of the random sample, the percentage of measurements that fall in the interval $\left[ 1.2, 2.4 \right]$ is $2.4 \%$.

The sample will not yield reliable values for a probability, the theoretical approximation and the actual calculation of the cumulative distribution function is the right way to find the probability (this cdf concept will become clear in further chapters).

**Code:**

```python

# Generate random normal sample with parameters
# This is a cell on its own to avoid regenerating the values again

women_hdl_sample = pd.Series(norm.rvs(0, 1.2, size = 2000), name = "hdl_difference", dtype = "float64" )

fg_hdl = sns.displot(women_hdl_sample, binwidth = 0.25, stat = "percent")
fg_hdl.set_axis_labels("HDL Difference (mg/L)", "Percent of Measurements")
fg_hdl.ax.set_title("Frequency of HDL difference for artificial sample of 2000 women.")
plt.show()

percentage_of_sample_measurements_2_5__3_6 = (women_hdl_sample[ (women_hdl_sample >= 2.5) & (women_hdl_sample <= 3.6)].count() / women_hdl_sample.count()) * 100
prob_2_5__3_6 = (norm.cdf( 3.6, loc = 0, scale = 1.2) - norm.cdf(2.5, loc = 0, scale = 1.2))*100

print(f"For this randomly generated sample the mean is {women_hdl_sample.mean():.4f} and the standard deviation is {women_hdl_sample.std():.4f} and the percentage of measurements of this sample in the interval  [2.5, 3.6] is {percentage_of_sample_measurements_2_5__3_6:.4f} %.")
print(f"For a normal distribution with population mean 0 mg/L and standard deviation 1.2 mg/L, the probability for the interval [2.5, 3.6] is {prob_2_5__3_6:.4f} %.")
print()


prob_1_2__2_4 = (norm.cdf(2.4, loc = 0, scale = 1.2) - norm.cdf(1.2, loc = 0, scale = 1.2)) * 100
percentage_of_sample_measurements_1_2__2_4 = (women_hdl_sample[ (women_hdl_sample >= 2.5) & (women_hdl_sample <= 3.6)].count() / women_hdl_sample.count()) * 100
print(f"Using the normal distribution of a population, the probability that a measurement is in the interval [1.2, 2.4] is {prob_1_2__2_4:.4f} %.")
print(f"For this dataset of 2000 measurements, the probability that a measurement is in the interval [1.2, 2.4] is {percentage_of_sample_measurements_1_2__2_4:.4f} %.")

```

### **Exercise 1.32**
#### **Statement:**

Let $k \geq 1$. Show that, for any set of $n$ measurements, the fraction included in the interval $\bar{y} - ks$ to $\bar{y} + ks$ is at least $\left( 1 - 1/k^{2} \right)$. [*Hint*: $$ s^{2} = \frac{1}{n-1} \left[ \sum_{i = 1}^{n}(y_{i} - \bar{y})^{2} \right].$$ In this expression, replace all deviations for which $|y_{i} - \bar{y}| \geq  ks$ with $ks$. Simplify.] This result is known as *Tchebysheff's theorem*. 

(Wackerly, Mendenhall & Scheaffer, 2008, p. 18)

#### **Solution**
**Theory Section:**

First, define the sets:

- $A := \left\{i \; | \; |y_{i} - \bar{y}| < ks \right\}$

- $B := \left\{i \; | \; |y_{i} - \bar{y}| \geq ks \right\}$

Set $A$ contains the values of $i$ where $y_{i}$ belongs to the interval $\left( \bar{y} - ks, \bar{y} + ks \right)$. The set $B$ contains the values of $i$ where $y_{i}$ belongs to the interval $\left( -\infty, \bar{y} - ks \right] \cup \left[ \bar{y} + ks, \infty  \right)$. Then,

\begin{align*}
s^{2} =& \frac{1}{n-1} \left[ \sum_{i=1}{n}\left( y_{i} - \bar{y} \right)^{2} \right] \\
\\
=& \frac{1}{n-1}\left[ \sum_{i \in A}\left( y_{i} - \bar{y} \right)^{2} + \sum_{i \in B}\left( y_{i} - \bar{y} \right)^{2} \right].
\end{align*}

Given that $|y_{i} - \bar{y}|\geq ks \; \; \Rightarrow \; \; (y_{i} - \bar{y})^{2} \geq (ks)^{2} $. This gives

\begin{align*}
s^{2} = \frac{1}{n-1}\left[ \sum_{i \in A}\left( y_{i} - \bar{y} \right)^{2} + \sum_{i \in B}\left( y_{i} - \bar{y} \right)^{2} \right] \geq & \; \frac{1}{n-1}\left[ \sum_{i \in A}\left( y_{i} - \bar{y} \right)^{2} + \sum_{i \in B}\left( ks \right)^{2} \right]
\end{align*}

By transitivity of inequalities its found that

\begin{align*}
s^{2} \geq & \; \frac{1}{n-1}\left[ \sum_{i \in A}\left( y_{i} - \bar{y} \right)^{2} + \sum_{i \in B}\left( ks \right)^{2} \right]
\end{align*}

The cardinality of $A$ and $B$ is $|A|$ and $|B|$, then

\begin{align*}
\frac{1}{n-1}\left[ \sum_{i \in A}\left( y_{i} - \bar{y} \right)^{2} + \sum_{i \in B}\left( ks \right)^{2} \right] = \frac{1}{n-1}\left[ \sum_{i \in A}\left( y_{i} - \bar{y} \right)^{2} + |B|\left( ks \right)^{2} \right]
\end{align*}

Knowing that for every $i$ in $A$, $(y_{i} - \bar{y}) ^{2} \geq 0$, then it follows that

\begin{align*}
s^{2} \geq \frac{1}{n-1}\left[ \sum_{i \in A}\left( y_{i} - \bar{y} \right)^{2} + |B|\left( ks \right)^{2} \right] \geq \frac{1}{n-1}|B|\left( ks \right)^{2}
\end{align*}

Now, given that $|B| = n - |A|$,

\begin{align*}
s^{2} \geq & \; \frac{1}{n-1}|B|\left( ks \right)^{2} \\
\\
\frac{n - 1}{k^{2}} \geq & \; |B| \\
\\
\frac{n - 1}{k^{2}} \geq & \; n -  |A| \\
\\
\frac{|A|}{n} \geq & \; 1 + \frac{1}{k^{2}n} - \frac{1}{k^{2}}
\end{align*}

Finally, with $\frac{1}{k^{2}n} \geq 0$

\begin{align*}
\frac{|A|}{n} \geq & \; 1 - \frac{1}{k^{2}}
\end{align*}

Here, $|A|/n$ is the fraction of values inside the interval $\left( \bar{y} - ks, \bar{y} + ks \right)$, it has $1 - 1/k^{2}$ as a lower bound.

Here $75\%$ of values is the proportion $0.75$ of all values, using the result from the previous exercise

\begin{align*}
\frac{|A|}{n} \geq 1 - \frac{1}{k^{2}}
\end{align*}

$k$ is in days. We are looking for the value of $k$ for which $\frac{|A|}{n} = 0.75$, as seen in the graph of the code section this is the intersection of the horizontal line at 0.75 and the function $f(k) =1-\frac{1}{k^{2}}$. Then, to find the value of $k$ that is the lower bound of the interval for which $\frac{|A|}{n} \geq 0.75$, the inequality is set to be

\begin{align*}
1 - \frac{1}{k^{2}} \geq 0.75
\end{align*}

Then

\begin{align*}
k \geq \sqrt{\frac{1}{1 - 0.75}} = 2
\end{align*}

The lower bound for $k$ is $2$. Then the interval to publish is $\left[ 0.5, 10.5 \right]$.

**Code:**

```python
# Ilustration of the lower bound.

k = np.linspace(1, 10, 500)
f_k = 1 - (1/(k*k))

fig, ax = plt.subplots()

sns.lineplot(x = k, y = f_k, ax = ax, label = r"$1 - \frac{1}{k^{2}}$" )
ax.hlines(0.75, xmin = 1, xmax = 10, color = "red", label = "Least proportion.")
ax.axvline(4, color = "red",linestyle = "--", label = r"Lower bound for $k$")

ax.set_xlabel("k")
ax.set_ylabel("Proportion")
fig.suptitle(r"$1 - \frac{1}{k^{2}}$ vs k")
ax.legend()
plt.show()
```

### **Exercise 1.34**
#### **Statement:**

For the data discussed in Exercise 1.33, give an upper bound to the fraction of days when there are more than 13 absentees.

(Wackerly, Mendenhall & Scheaffer, 2008, p. 18)

#### **Solution**
**Theory Section:**

First, find a $k$ such that the interval $\left( 13 , \infty \right)$ is expressed in terms of k. Using the expression $13 = 5.5 + k(2.5)$, then, $k = \frac{13 - 5.5}{2.5} = 3$, this means that a day with 13 absentees is 3 times the standard deviation from the mean. By Tchebysheff's theorem, the fraction contained inside the interval is 

\begin{align*}
\frac{|A|}{n} \geq & \; 1 - \frac{1}{k^{2}} \\
\\
\frac{|A|}{n} \geq & \; 1 - \frac{1}{3^{2}} \\
\\
\frac{|A|}{n} \geq & \; \frac{8}{9}
\end{align*}

By definition, $\frac{|A|}{n} = 1 - \frac{|B|}{n}$,

\begin{align*}
1 - \frac{|B|}{n} \geq & \; \frac{8}{9}\\
\\
\frac{|B|}{n} \leq & \; 1 - \frac{8}{9}\\
\\
\frac{|B|}{n} \leq & \; \frac{1}{9}
\end{align*}

This means that at most 1/9 of the days can have 13 or more absentees.

1. Approximate the value of $s$ using the range approximation.

2. Calculate the values of $\bar{y}$ and $s$ for the 15 blood pressure readings.

3. Use Tchebysheff's theorem (Exercise 1.32) to find values $a$ and $b$ such that at least $75\%$ of the blood pressure measurements lie between $a$ and $b$.

4. Did Tchebysheff's theorem work? That is, use the data to find the actual percent of blood pressure readings that are between the values $a$ and $b$ you found in part (3). Is this actual percentage greater than $75\%$?

(Wackerly, Mendenhall & Scheaffer, 2008, p. 18)

#### **Solution**
**Theory Section:**

1. The range approximation for $s$ states that the $s$ of a set of measurements may be roughly approximated by one-fourth of the range. This is by taking the largest and smallest value from the blood pressure recordings and finding the range
\begin{align*}
s \approx \frac{\mathrm{range}}{4} = \frac{172 - 108}{4} = 16
\end{align*}

2. By finding the values using pandas it can be seen that $\bar{y} = 136.0667$ and $s=17.1025$.

3. To find the values $a$ and $b$ for which at least $75\%$ of the measurements are, that is $\frac{|A|}{n} \geq 0.75$, we use the result of exercise 1.32

    \begin{align*}
    0.75 & \leq 1 - \frac{1}{k^{2}} \\
    1 - 0.75 & \geq \frac{1}{k^{2}} \\
    k & \geq \sqrt{\frac{1}{1 - 0.75}} \\
    k & \geq 2
    \end{align*}

    $k$ has to be at least 2. Then, one can find the  values $a = \bar{y} - ks = 101.8617$ and $b = \bar{y} + ks = 170.2717$. The interval that contains at least $75\%$ of the measurements is $\left[101.8617, 170.2717 \right]$.

4. Using a python script and filtering the data frame one can find that $93.33\%$ of the values are inside this interval which is larger than $75\%$ the least proportion of values that should be found.

**Code:**

```python
# ==========================> Answer to 2. <===============================

raw_data = [172, 140, 123, 130, 115,
            148, 108, 129, 137, 161,
            123, 152, 133, 128, 142]

blood_pressures = pd.Series(raw_data, name="blood_pressure", dtype = "float64[pyarrow]")

print(f"The mean of the measurements is {blood_pressures.mean():.4f} and the standard deviation is {blood_pressures.std():.4f} both in the units of measurement.")

# ==========================> Answer to 4. <===============================

true_fraction = blood_pressures[(blood_pressures >= 101.8617) & (blood_pressures <= 170.2717)].count()/blood_pressures.count()

print(f"The amount of values that are inside is {true_fraction * 100:.4f} %.")
```

### **Exercise 1.36**
#### **Statement:**

A random sample of 100 foxes was examined by a team of veterinarians to determine the prevalence of a specific parasite. Counting the number of parasites of this specific type, the veterinarians found that 69 foxes had no parasites of the type of interest, 17 had one parasite of the type under study, and so on. A summary of their results is given in the following table

<center><img src="img/im5.png" alt="im4" width="400px"/></center>

1. Construct the relative frequency histogram for the number of parasites per fox.

2. Calculate $\bar{y}$ and $s$ for the data given.

3. What fraction of the parasite counts falls within 2 standard deviations of the mean? Within 4 standard deviations? Do your results agree with Tchebysheff's theorem (exercise 1.32) and/or the empirical rule?

(Wackerly, Mendenhall & Scheaffer, 2008, p. 18, 19)

#### **Solution**
**Theory Section:**

To solve this, some python implementation is necessary.

1. Below is the code to implement the histogram for the given prevalence of parasites.

2. Using pandas vectorization it can be found that $\bar{y} = 0.66$ and $s = 1.3870$.

3. First, using pandas it can be found that for the interval $\left[ 0, 3.434 \right]$ the fraction of values that fall within this interval is 0.95. (the left limit is 0 since $\bar{y}-2s = -2.114$ has no meaning for this problem). For Tchebysheff's theorem we have that $k=2$, then $1 - 1/2^{2} = 0.75$ which is correct, at least, the fraction 0.75 of the measurements is between $\left[ 0, 3.434 \right]$, then the empirical rule means that $95\%$ of values is contained in this interval which is also correct, but for this kind of distribution it makes no sense.

**Code:**

```python
# ==========================> Answer to 1. <===============================

raw_data = [69, 17, 6, 3, 1, 2, 1, 0, 1]

parasites_series = pd.Series(raw_data, name = "number_of_parasites", dtype = "float64[pyarrow]")
parasites_series_relative_frequency = parasites_series / parasites_series.sum() * 100

fig, ax = plt.subplots()
sns.barplot(x = parasites_series_relative_frequency.index, y = parasites_series_relative_frequency, ax = ax)
#parasites_series_relative_frequency.plot.bar(ax = ax, rot = 0)
ax.set_title("Relative Frequency for the Prevalence of a Specific Parasite in Foxes")
ax.set_xlabel("Number of Parasites")
ax.set_ylabel("Relative Frequency")
plt.show()

# ==========================> Answer to 2. <===============================

mean_parasites = (parasites_series.index * parasites_series).sum() / parasites_series.sum()
std_parasites  = np.sqrt((parasites_series * ((parasites_series.index - mean_parasites) ** 2) ).sum() / (parasites_series.sum() - 1))
print(f"The mean is {mean_parasites:.4f} and the standard deviation is {std_parasites:.4f}")

# ==========================> Answer to 3. <===============================

fraction_2s_parasites = parasites_series[(parasites_series.index >= 0) & (parasites_series.index <= 3.434)].sum() / parasites_series.sum()
print(f"The fraction of values that fall in the interval [0, 3.434] is {fraction_2s_parasites:.4f}.")
```
